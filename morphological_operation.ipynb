```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
!curl -o logo.png https://colab.research.google.com/img/colab_favicon_256px.png
import cv2
img = cv2.imread('/content/Neuron.jpg')
imgray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

_,mask=cv2.threshold(imgray,25,255,cv2.THRESH_BINARY)

plt.subplot(1,3,1)
plt.imshow(img,cmap='gray')
plt.title('original image')
plt.subplot(1,3,2)
plt.imshow(imgray,cmap='gray')
plt.title('gray image')
plt.subplot(1,3,3)
plt.imshow(mask,cmap='gray')
plt.title('mask')
plt.show()
```

erosion

```python
kernel=np.ones((3,3),np.uint8)
eroded_img=cv2.erode(mask,kernel,iterations=1)
eroded_img1=cv2.erode(mask,(5,5),iterations=1)

plt.subplot(1,4,1)
plt.imshow(imgray,cmap='gray')
plt.title('gray image')
plt.subplot(1,4,2)
plt.imshow(mask,cmap='gray')
plt.title('mask')
plt.subplot(1,4,3)
plt.imshow(eroded_img,cmap='gray')
plt.title(' eroded_img')
plt.subplot(1,4,4)
plt.imshow(eroded_img1,cmap='gray')
plt.title(' eroded_img1')
plt.show()
```

dilation

```python


kernel=np.ones((3,3),np.uint8)
deliated_img=cv2.dilate(mask,kernel,iterations=1)
eroded_img=cv2.erode(deliated_img,kernel,iterations=1)


plt.subplot(1,4,1)
plt.imshow(imgray,cmap='gray')
plt.title('gray image')
plt.subplot(1,4,2)
plt.imshow(mask,cmap='gray')
plt.title('mask')
plt.subplot(1,4,3)
plt.imshow(deliated_img,cmap='gray')
plt.title(' deliated_img')
plt.subplot(1,4,4)
plt.imshow(eroded_img,cmap='gray')
plt.title(' eroded_img')
plt.show()
```

closing

```python
img1 = cv2.imread('/content/cells_threshold.png')
imgray1=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)

_,mask1=cv2.threshold(imgray1,25,255,cv2.THRESH_BINARY)
```

```python
kernel=np.ones((3,3),np.uint8)
closed_img=cv2.morphologyEx(mask1,cv2.MORPH_CLOSE,kernel)

kernel2=np.ones((5,5),np.uint8)
closed_img2=cv2.morphologyEx(mask1,cv2.MORPH_CLOSE,kernel2)

plt.subplot(1,4,1)
plt.imshow(imgray1,cmap='gray')
plt.title('gray image')
plt.subplot(1,4,2)
plt.imshow(mask1,cmap='gray')
plt.title('mask')
plt.subplot(1,4,3)
plt.imshow(closed_img,cmap='gray')
plt.title(' closed_img')
plt.subplot(1,4,4)
plt.imshow(closed_img2,cmap='gray')
plt.title(' closed_img2')
plt.show()
```

opening

```python
kernel=np.ones((9,9),np.uint8)
opened_img=cv2.morphologyEx(mask1,cv2.MORPH_OPEN,kernel)

kernel2=np.ones((25,25),np.uint8)
opened_img2=cv2.morphologyEx(mask1,cv2.MORPH_OPEN,kernel2)

plt.subplot(1,4,1)
plt.imshow(imgray1,cmap='gray')
plt.title('gray image')
plt.subplot(1,4,2)
plt.imshow(mask1,cmap='gray')
plt.title('mask')
plt.subplot(1,4,3)
plt.imshow(opened_img,cmap='gray')
plt.title(' opened_img')
plt.subplot(1,4,4)
plt.imshow(opened_img2,cmap='gray')
plt.title(' opened_img2')
plt.show()
```

Gradient

```python
kernel=np.ones((5,5),np.uint8)
gradient=cv2.morphologyEx(mask1,cv2.MORPH_GRADIENT,kernel)

plt.subplot(1,2,1)
plt.imshow(mask1,cmap='gray')
plt.title('mask1')
plt.subplot(1,2,2)
plt.imshow(gradient,cmap='gray')
plt.title('gradient')
```

```python
kernel2 = np.ones((5,5), np.uint8)
delated_img = cv2.dilate(mask1, kernel2, iterations=1)
subtract_img = delated_img - mask1


plt.subplot(1,3,1)
plt.imshow(mask1,cmap='gray')
plt.title('mask1')
plt.subplot(1,3,2)
plt.imshow(subtract_img,cmap='gray')
plt.title('subtract_img')
plt.subplot(1,3,3)
plt.imshow(delated_img,cmap='gray')
plt.title('delated_img')
```
