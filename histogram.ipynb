```python
#histogram for color image
from google.colab.patches import cv2_imshow
!curl -o logo.png https://colab.research.google.com/img/colab_favicon_256px.png
import cv2
img = cv2.imread('/content/4seasons.jpg')
colors=['b','g','r']
for i,col in enumerate(colors):
  histogram=cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(histogram,color=col)
plt.show()
```

```python
#histogram for color image with mask
from google.colab.patches import cv2_imshow
!curl -o logo.png https://colab.research.google.com/img/colab_favicon_256px.png
import cv2
img = cv2.imread('/content/4seasons.jpg')
w,h,c=img.shape
mask=np.zeros(img.shape[0:2],np.uint8)
mask[0:int(w/2),0:int(h/2)]=255
colors=['b','g','r']
for i,col in enumerate(colors):
  histogram=cv2.calcHist([img],[i],mask,[256],[0,256])
  plt.plot(histogram,color=col)
plt.show()
```

```python
from google.colab.patches import cv2_imshow
!curl -o logo.png https://colab.research.google.com/img/colab_favicon_256px.png
import cv2
img = cv2.imread('/content/statue.jpg',0)

histogram=cv2.calcHist([img],[0],None,[256],[0,256])

#histogram equalization
equalized_histogram=cv2.equalizeHist(img)
img_equal_hist=cv2.calcHist([equalized_histogram],[0],None,[256],[0,256])

#adaptive histogram equalization
clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
cl1=clahe.apply(img)
cl_equal_hist=cv2.calcHist([cl1],[0],None,[256],[0,256])

plt.subplot(321)
plt.imshow(img,'gray')
plt.title('img')
plt.subplot(322)
plt.plot(histogram)
plt.title('histogram')

plt.subplot(323)
plt.imshow(equalized_histogram,'gray')
plt.title('equalized_histogram')
plt.subplot(324)
plt.plot(img_equal_hist)
plt.title('eualized_histogram')

plt.subplot(325)
plt.imshow(cl1,'gray')
plt.title('clahe')
plt.subplot(326)
plt.plot(cl_equal_hist)
plt.title('clahe')
plt.show()
```
