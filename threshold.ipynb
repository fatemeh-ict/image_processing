```python
import numpy as np
import cv2
```

```python
from google.colab.patches import cv2_imshow
!curl -o logo.png https://colab.research.google.com/img/colab_favicon_256px.png
import cv2
img = cv2.imread('/content/book.jpg')
imgray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,threshold=cv2.threshold(img,10,255,cv2.THRESH_BINARY)
ret,threshold1=cv2.threshold(imgray,10,255,cv2.THRESH_BINARY)

cv2_imshow(img)
cv2_imshow(threshold)
cv2_imshow(threshold1)
```

```python
img.shape
```

```python
#OTSU Threshold
imgray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret2,threshold2=cv2.threshold(imgray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print(ret2)
cv2_imshow(threshold2)
```

```python
#Adaptive Threshold
threshold3=cv2.adaptiveThreshold(imgray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,115,1)
cv2_imshow(threshold3)
```

```python
img = cv2.imread('/content/brain.jpeg')
imgray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,threshold=cv2.threshold(img,50,255,cv2.THRESH_BINARY)
ret,threshold1=cv2.threshold(imgray,80,255,cv2.THRESH_BINARY)

cv2_imshow(img)
cv2_imshow(threshold)
cv2_imshow(threshold1)
```

```python
#OTSU Threshold
imgray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret2,threshold2=cv2.threshold(imgray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print(ret2)
cv2_imshow(threshold2)
```

```python
#Adaptive Threshold
threshold3=cv2.adaptiveThreshold(imgray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,15,1)
cv2_imshow(threshold3)
```
