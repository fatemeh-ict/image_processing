```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('/content/jumbo.jpg')

plt.imshow(img)
plt.show()
```

#Sift

```python
feat_sift=cv2.xfeatures2d.SIFT_create()
sift_keypoints,descriptor=feat_sift.detectAndCompute(img,None)
img_sift=cv2.drawKeypoints(img,sift_keypoints,None)

plt.imshow(img_sift)
plt.show()
```

#ORB

```python
feat_orb=cv2.ORB_create(nfeatures=1000)
orb_keypoints,descriptors=feat_orb.detectAndCompute(img,None)
img_orb=cv2.drawKeypoints(img,orb_keypoints,None)

plt.imshow(img_orb)
plt.show()
```

```python
img1=cv2.imread('/content/book1.jpeg')
img2=cv2.imread('/content/book2.jpeg')

```

```python
#orb
feat_orb=cv2.ORB_create(1000)
orb_keypoints1,descriptors1=feat_orb.detectAndCompute(img1,None)
orb_keypoints2,descriptors2=feat_orb.detectAndCompute(img2,None)

bf=cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)

matches=bf.match(descriptors1,descriptors2)
matches=sorted(matches,key=lambda x:x.distance)

img_match=cv2.drawMatches(img1,orb_keypoints1,img2,orb_keypoints2,matches[:100],None,flags=2)

plt.imshow(img_match)
plt.show()
```
