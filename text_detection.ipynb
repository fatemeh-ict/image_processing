```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import easyocr
```

```python
car_img=cv.imread('/content/Car_plate_en2.jpg')
car_plate_img = cv.imread('/content/plate_en2_2.jpg',0)
plt.imshow(cv.cvtColor(car_plate_img,cv.COLOR_BGR2RGB))
plt.show()
```

```python
reader=easyocr.Reader(['en'])
plate_info=reader.readtext(car_plate_img)
print(plate_info)
```

```python
plate_text=plate_info[0][-2]
print(plate_text)
```
