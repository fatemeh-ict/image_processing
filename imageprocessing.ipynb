```python
import numpy as np
import cv2
```

```python

!curl -o logo.png https://colab.research.google.com/img/colab_favicon_256px.png
img = cv2.imread('/content/cityscape.jpg', cv2.IMREAD_UNCHANGED)
cv2.line(img,(5,5),(200,150),(200,200,0),10)
cv2.rectangle(img,(5,5),(200,150),(255,255,255),6)
cv2.circle(img,(200,150),50,(0,255,255),-1)
cv2.circle(img,(75,75),50,(0,255,255),20)
points=np.array([[200,200],[250,250],[300,300],[225,150],[275,175]],np.int32)
cv2.polylines(img,[points],True,(255,0,0),5)
cv2_imshow(img)
```

```python
img = cv2.imread('/content/cityscape.jpg', cv2.IMREAD_UNCHANGED)
font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,'Hello World',(600,100),font,1,(255,255,255),2,cv2.LINE_AA)
cv2_imshow(img)
```
