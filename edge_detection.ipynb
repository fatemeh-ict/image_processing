```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
!curl -o logo.png https://colab.research.google.com/img/colab_favicon_256px.png
import cv2
img = cv2.imread('/content/Road_in_Norway.jpg')
imgray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
cv2_imshow(imgray)
```

#laplacian

```python
image_noise_removed=cv2.GaussianBlur(imgray,(5,5),0)
laplacian1=cv2.Laplacian(image_noise_removed,cv2.CV_64F)

plt.subplot(1,2,1)
plt.imshow(image_noise_removed,cmap='gray')
plt.title('blure image')
plt.subplot(1,2,2)
plt.imshow(laplacian1,cmap='gray')
plt.title('laplacian1')
plt.show()
```

```python
laplacian=cv2.Laplacian(imgray,cv2.CV_64F)

plt.subplot(1,2,1)
plt.imshow(imgray,cmap='gray')
plt.title('image gray')
plt.subplot(1,2,2)
plt.imshow(laplacian,cmap='gray')
plt.title('laplacian')
plt.show()
```

Sobel X

```python
sobel_x=cv2.Sobel(imgray,cv2.CV_64F,1,0,ksize=5)
sobel_x1=cv2.Sobel(image_noise_removed,cv2.CV_64F,1,0,ksize=5)

plt.subplot(1,2,1)
plt.imshow(sobel_x,cmap='gray')
plt.title('with imgray')
plt.subplot(1,2,2)
plt.imshow(sobel_x1,cmap='gray')
plt.title('with blured image')
plt.show()
```

#Sobel Y

```python
sobel_y=cv2.Sobel(imgray,cv2.CV_64F,0,1,ksize=5)
sobel_y1=cv2.Sobel(image_noise_removed,cv2.CV_64F,0,1,ksize=5)

plt.subplot(1,2,1)
plt.imshow(sobel_y,cmap='gray')
plt.title('with imgray')
plt.subplot(1,2,2)
plt.imshow(sobel_y1,cmap='gray')
plt.title('with blured image')
plt.show()
```

Canny

```python
canny=cv2.Canny(imgray,100,200)
canny1=cv2.Canny(image_noise_removed,100,200)

plt.subplot(1,2,1)
plt.imshow(canny,cmap='gray')
plt.title('with gray image')
plt.subplot(1,2,2)
plt.imshow(canny1,cmap='gray')
plt.title('with blured image')
plt.show()
```

```python
#without removing noise
plt.subplot(1,5,1)
plt.imshow(imgray,cmap='gray')
plt.title('original')
plt.subplot(1,5,2)
plt.imshow(laplacian,cmap='gray')
plt.title('laplacian')
plt.subplot(1,5,3)
plt.imshow(sobel_x,cmap='gray')
plt.title('sobel_x')
plt.subplot(1,5,4)
plt.imshow(sobel_y,cmap='gray')
plt.title('sobel_y')
plt.subplot(1,5,5)
plt.imshow(canny,cmap='gray')
plt.title('canny')
```
