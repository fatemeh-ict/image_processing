```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
!curl -o logo.png https://colab.research.google.com/img/colab_favicon_256px.png
import cv2
img = cv2.imread('/content/planes.jpeg',0)
template=cv2.imread('/content/plane.jpeg',0)


cv2_imshow(img)
cv2_imshow(template)
```

```python
results=cv2.matchTemplate(img,template,cv2.TM_CCOEFF_NORMED)
plt.imshow(results)
plt.show()
```

```python
w,h=template.shape
threshold=.5
location=np.where(results>=threshold)
for point in zip(*location[::-1]):
  cv2.rectangle(img,point,(point[0]+h,point[1]+w),(255,255,0),2)
plt.imshow(img)
plt.show()
```
